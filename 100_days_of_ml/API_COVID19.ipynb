```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
response = requests.get("https://api.openweathermap.org/data/2.5/forecast?id=524901&appid=1122bcc21c88f8ef322db806dee08dbf")
```

```python
import json
data = response.json()
```

```python
print(json.dumps(data,indent=2)) ## Basically prints the JSON View
```

```python
temp_df = pd.json_normalize(data['list'])  # EXTRACT and normalize"
temp_df
```

```python
temp_df.shape
```

```python
# Extract main weather description from the first item in the weather list
temp_df['weather_main'] = temp_df['weather'].apply(lambda x: x[0]['main'] if isinstance(x, list) and len(x) > 0 else None)
temp_df['weather_description'] = temp_df['weather'].apply(lambda x: x[0]['description'] if isinstance(x, list) and len(x) > 0 else None)

# Drop the messy original 'weather' column if you don't need it
temp_df = temp_df.drop(columns=['weather'])
```
