# Numerical - Numerical


# Scatter Plot


```python
sns.scatterplot(x = tips["total_bill"] , y= tips["tip"] , data =tips , hue = tips["sex"] , style= tips["smoker"] , size = tips["size"])
```

# BAR PLOT (NUMERICAL - CATEGORICAL )


```python
sns.barplot(x="Pclass" , y= "Age" , data = df , hue = "Sex" )
```

# BOX PLOT (Numerical - Categorical)


```python
df.head()
```

```python
sns.boxplot(x="Sex" , y= "Age" , data=df , hue="Survived")
```

# DIST PLOT(Numerical - Categorical)

```python
sns.kdeplot(df[df["Survived"] == 1]["Age"], label="Survived", shade=True)
sns.kdeplot(df[df["Survived"] == 0]["Age"], label="Not Survived", shade=True)
```

# HEATMAP (Categorical - CAtegorical)

```python
pd.crosstab(df["Pclass"],df["Sex"])
```

```python
sns.heatmap(pd.crosstab(df["Pclass"],df["Sex"]) )
```

```python

df["Survived"] = pd.to_numeric(df["Survived"] , errors = "coerce")
df.groupby("Pclass")["Survived"].mean()*100
```

```python
df.groupby("Sex")["Survived"].mean()*100
```

```python
df.groupby("Embarked")["Survived"].mean()*100
```

# ClusterMap (Cat-Cat)


```python
sns.clustermap(pd.crosstab(df["SibSp"] ,df["Survived"]))
```
